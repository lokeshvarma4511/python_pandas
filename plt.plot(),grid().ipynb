# having multiple graphs on a single one


```python

from matplotlib import style
days=[1,2,3,4,5,7,8,9,10,11,13,14,15,16,17]
jan_temp=[25,22,26,20.6,18.3,17,23.6,20,29,27,23.1,-3,16.7,12,30]
feb_temp=[22,17,26,29,20.6,25,23.1,18.3,30,23.6,20,27,-3,16.7,12]

style.use("ggplot")
plt.axis([0,18, -5,30])
plt.plot(days,jan_temp,"go-.",linewidth=2,markersize=8,label='jan_temp')
plt.plot(days,feb_temp,"yo-.",linewidth=2,markersize=8,label='feb_temp')

plt.title("january & february temperatures")
plt.xlabel("days")
plt.ylabel("jan & feb Temperature")
plt.legend(loc=4)
plt.grid(color='k', linestyle='-',linewidth=2)
plt.show()
      
```
