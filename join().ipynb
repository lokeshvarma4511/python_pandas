```python
import pandas as pd
```

```python
df3=pd.DataFrame({'I':[31,2,69,4,50],'Grad':['A','B','C','D','E'],'Na':['ab','ac','ad','ae','af']},index=[1,2,3,4,5])
df4=pd.DataFrame({'ID':[23,6,562,256],'Grade':['B','C','D','E'],'Name':['pass','gh','so','sum']},index=['a','b','c','d'])
```

```python
df3,df4
```

```python
df3.join(df4)
```

```python
df4.join(df3)
```

if we want to add 2 differently possed DF's index 

```python
df3.join(df4,how='inner') #intersection indexes
```

```python
df3.join(df4,how='left')
```

```python
df3.join(df4,how='right')
```

```python
df3.join(df4,how='outer')    #union
```

```python
df3=pd.DataFrame({'I':[31,2,69,4,50],'Grad':['A','B','C','D','E'],'Na':['ab','ac','ad','ae','af']},index=[1,2,3,4,5])
df4=pd.DataFrame({'ID':[23,6,562,256],'Name':['pass','gh','so','sum'],'Grad':['B','C','D','E'],},index=['a','b','c','d'])
```

```python
df3.join(df4,how='outer',lsuffix='_l',rsuffix='_r')
```

```python
df3.join(df4,lsuffix='_l',rsuffix='_r',sort=True)
```
