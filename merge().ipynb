```python
df1=pd.DataFrame({'ID':[1,2,3,4,8],'Class':[6,7,8,9,10]})
```

```python
df2=pd.DataFrame({'ID':[1,2,3,4,5],'Name':['A','B','C','D','E']})
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,how='inner')
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1,df2,how='left')
```

```python
pd.merge(df1,df2,how='right')
```

```python
pd.merge(df1,df2,how='outer')
```

```python
pd.merge(df1,df2,how='outer',indicator='True')
```

if we want to merge both same columns into one df

```python
pd.merge(df1,df2,left_index=True,right_index=True)
```

```python
pd.merge(df1,df2,on='ID',)
```

```python
df3=pd.DataFrame({'ID':[1,2,3,4,5],'Name':['A','B','C','D','E']})
```

```python
df4=pd.DataFrame({'ID':[1,2,3,4,5],'Name':['A','B','C','D','E']})
```

```python
pd.merge(df3,df4,on='ID')
```

```python
pd.merge(df3,df4,on=['ID','Name'])
```

```python
pd.merge(df3,df4,left_index=True,right_index=True)
```

if we want to change the names of the the same columns then use         suffixes=(' ',' ')

```python
pd.merge(df3,df4,on='ID',suffixes=('_first','_then'))
```
